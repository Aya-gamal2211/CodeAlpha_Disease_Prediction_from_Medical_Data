Find Outliers

```python
Q1=train_data.quantile(0.25)
Q3=train_data.quantile(0.75)

IQR= Q3-Q1

#Get the whiskers
min= Q1 -(1.5*IQR)
max= Q3 +(1.5*IQR)

outliers=train_data[(train_data < min) | (train_data> max)]

print(outliers)
```
